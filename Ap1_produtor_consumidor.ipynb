```python
# Importações
import threading
import time
import random
```

```python
class bcolors:
    green = '\033[92m' #GREEN
    yellow = '\033[93m' #YELLOW
    red = '\033[91m' #RED
    RESET = '\033[0m' #RESET COLOR
```

```python
# Função produtor
def produtor(itens):
    i = 0
    while True:
        time.sleep(random.uniform(2, 3))
        itens.append(i)
        print(bcolors.green + "Produziu: "+str(i) + bcolors.RESET)
        i+=1
```

```python
# Função consumidor
def consumidor(itens):
    while True:
        time.sleep(random.uniform(2, 3))
        if len(itens) == 0:
            print(bcolors.yellow + "Sem itens à consumir" + bcolors.RESET)
            continue
        item = itens.pop(0)
        print(bcolors.red + "Consumiu: "+str(item) + bcolors.RESET)
        print(len(itens))
```

```python
# Criação de Threads
itens = []
producer = threading.Thread(target=produtor,args=(itens,))
consumer = threading.Thread(target=consumidor,args=(itens,))
```

```python
producer.start()
consumer.start()
```
